# Air Pollution is Rising!
## Predicting Pollution Levels using Machine Learning
It is winter time in Delhi so Cody decided to go for a walk to the news stand and on reaching
was surprised to see the air quality index of Delhi in the newspaper . So he decided to collect
the air samples of different locations and then took these samples to his lab where he extracted
five features of the air he collected which can be used to predict the air quality index and
combined it with the air quality index given in the newspapers. You are provided with the data
collected by Cody and your job is to design a machine learning model which given the features
extracted by Cody can predict air quality.

## Data consist of three csv files :-
1. Train.csv
2. Test.csv

Train:- The train data consists of five feature columns and one target column. All the feature
columns are numeric and target value is also numeric.

Test:- Test data consist all the columns contained by train with the exception of target column
which u have to predict using the model trained on train data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_data = pd.read_csv('Train.csv')
train_data
```

```python
x1=train_data['feature_1'].values.reshape((-1,1))
x2=train_data['feature_2'].values.reshape((-1,1))
x3=train_data['feature_3'].values.reshape((-1,1))
x4=train_data['feature_4'].values.reshape((-1,1))
x5=train_data['feature_5'].values.reshape((-1,1))
x=train_data[['feature_1','feature_2','feature_3','feature_4','feature_5']].values
y=train_data['target'].values.reshape((-1,1))
```

```python
plt.figure(figsize=(10,10))
plt.scatter(x1,y,label='feature_1')
plt.scatter(x2,y,label='feature_2')
plt.scatter(x3,y,label='feature_3')
plt.scatter(x4,y,label='feature_4')
plt.scatter(x5,y,label='feature_5',color='yellow')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
output = model.predict(x)
```

```python
model.score(x,y)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x1,y,label='feature_1')
plt.scatter(x2,y,label='feature_2')
plt.scatter(x3,y,label='feature_3')
plt.scatter(x4,y,label='feature_4')
plt.scatter(x5,y,label='feature_5',color='yellow')
plt.plot(x,output,label='predicted',color='orange')
plt.legend()
plt.show()
```

```python
test_data=pd.read_csv('Test.csv')
test_data
```

```python
x1_test = test_data['feature_1'].values.reshape((-1,1))
x2_test = test_data['feature_2'].values.reshape((-1,1))
x3_test = test_data['feature_3'].values.reshape((-1,1))
x4_test = test_data['feature_4'].values.reshape((-1,1))
x5_test = test_data['feature_5'].values.reshape((-1,1))
t=test_data[['feature_1','feature_2','feature_3','feature_4','feature_5']].values
print(t,t.shape,type(t),sep='\n')
```

```python
testoutput = model.predict(t)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x1_test,testoutput,label='feature_1')
plt.scatter(x2_test,testoutput,color="yellow",label='feature_2')
plt.scatter(x3_test,testoutput,color="orange",label='feature_3')
plt.scatter(x4_test,testoutput,color="blue",label='feature_4')
plt.scatter(x5_test,testoutput,color="pink",label='feature_5')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
plt.close()
```

```python

plt.figure(figsize=(10,10))
plt.plot(t,testoutput,label='predictons')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
predicted_output = test_data
predicted_output['target_predicted']=testoutput
predicted_output
```
